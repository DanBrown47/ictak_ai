```python
# Loading libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv('data.csv')
```

```python
df.head() # checking if we got the data right
```

```python
# Here on inspection, There is multiple NaN values, Need to fill them using fillna()
# Procuring the mean values for Calories
calories_mean = df['Calories'].mean()
df['Calories'] = df['Calories'].fillna(calories_mean)
```

```python
# The same needs to be done for  the Date, here we can use the interpolate to do so
# The data gets converted to datetime
# Then using the Forward Filling function
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].ffill(inplace=True)
```

```python
# setting up date as index 
df.set_index('Date', inplace=True)
```

```python
plt.subplot(1, 2, 1)
sns.lineplot(data=df, x=df.index, y='Pulse')
sns.lineplot(data=df, x=df.index, y='Maxpulse')
plt.xlabel('Date')
plt.ylabel('Pulse')
plt.title('Pulse and Maxpulse')


plt.subplot(1, 2, 2)
sns.lineplot(data=df, x=df.index, y='Calories')
sns.lineplot(data=df, x=df.index, y='Duration')
plt.xlabel('Date')
plt.ylabel('Calories and Duration')
plt.title('Calories and Duration')


plt.show()
```
